```python
import csv
import requests
from lxml import html
import string
import os

alphabets = string.ascii_lowercase
url = 'http://www.economist.com/economics-a-to-z/'
# User-Agent to fake browser access
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'}

def scrapEconomist(url):
    page = requests.get(url, headers=header)
    return html.fromstring(page.content)
def get_term_h2(tree):
    return tree.xpath('//div[@class="item-list"]//h2/text()')
def get_defintion_div(tree):
    definition =tree.xpath('//li//div[@class="content clearfix"]//text()')
    defList = []
    i = -1
    for d in definition:
        
        if d in ['\n    ']:
            i += 1
            defList.append('')
        if d not in ['\n  ']:
            defList[i] = defList[i] + d + ' '
    return defList
```

```python
myDict = {}
for a in alphabets:
    web = url + a   
    tree = scrapEconomist(web)
    terms = get_term_h2(tree)
    definitions = get_defintion_div(tree)             
    if len(terms) == len(definitions):
        myDict.update({ a.upper()  : dict(zip(terms, definitions))})
#        print(myDict)
    else:
        print(a)

```

```python
def write_def_to_csv(file, data):
    '''
    Continue to write to csv file w/o overwriting exisitng text
    '''
    try:        
        with open(file, 'a') as f:
            writer = csv.writer(f)
            writer.writerows(data)
        
    except IOError as err:
        print(err)
    return

def read_econ_csv(file):
    with open(file, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            print(row)
```

```python
path = os.getcwd() + '/dict.csv'
print(path)
# write a header to csv file
with open(path, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['Alphabet','Term','Definition'])
    
for a in alphabets.upper():
    rows = []
    for k,v in myDict[a].items():        
        rows.append([a, k, v])
        if len(rows) == len(myDict[a]):
            write_def_to_csv(path, rows)
            print('Done write ' + a)


```
